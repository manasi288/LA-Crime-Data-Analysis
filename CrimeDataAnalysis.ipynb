```python
crimeDataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IE 6600/Project/Crime_Data_from_2020_to_Present.csv")
crimeDataset
```

```python
crimeDataset['Vict Sex'] = crimeDataset['Vict Sex'].fillna('Unknown')
crimeDataset['Premis Cd'] = crimeDataset['Premis Cd'].fillna('Unknown')
crimeDataset.isnull().sum()
```

```python
removeColumn = ['Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4', 'Cross Street' ]
crimeDataset.drop(removeColumn, inplace = True, axis = 1)
crimeDataset.isnull().sum()
```

```python
crimeDataset = crimeDataset.dropna(subset=['Mocodes', 'Premis Desc', 'Vict Descent', 'Weapon Used Cd', 'Weapon Desc'])
print(f" The total number of Rows is: {len(crimeDataset)}")
print(crimeDataset.isnull().sum())
```

```python
columnName =  {'DR_NO': 'Division of Record No.',
               'Date Rptd':'Date Reported',
               'Rpt Dist No': 'Reported District No',
               'Crm Cd': 'Crime Code',
               'DATE OCC': 'Date of Occurrence',
               'TIME OCC' : 'Time of Occurrence',
               'Crm Cd Desc': 'Crime Code Description',
               'Vict Age': 'Victim Age',
               'Vict Sex': 'Victim Sex',
               'AREA' : 'Area Number',
               'AREA NAME' : 'Area Name',
               'Part 1-2': 'Crime Type- Part 1/2',
               'Vict Descent': 'Victim Descent',
               'Premis Cd': 'Premis Code',
               'Premis Cd': 'Premis Description',
               'Weapon Used Cd': 'Weapon Used Code',
               'Weapon Desc': 'Weapon Description',
               'LOCATION' : 'Location',
               'Status Desc ': 'Status Desccription ',
               'LAT': 'Latitude','LON': 'Longitude'}
crimeDataset.rename(columns=columnName, inplace=True)
crimeDataset
```

```python
crimeDataset['Victim Sex'] = crimeDataset['Victim Sex'].replace("H", "X")
```

```python
crimeDataset['Date Reported'] = pd.to_datetime(crimeDataset['Date Reported'])
crimeDataset['Date of Occurrence'] = pd.to_datetime(crimeDataset['Date of Occurrence'])
```

```python
crimeDataset['Date Reported'] = crimeDataset['Date Reported'].dt.date
crimeDataset['Date of Occurrence'] = crimeDataset['Date of Occurrence'].dt.date
```

```python
crimeDataset['Time of Occurrence'] = crimeDataset['Time of Occurrence'].astype(str).str.zfill(4).str[:-2] + ':' + crimeDataset['Time of Occurrence'].astype(str).str.zfill(4).str[-2:]
crimeDataset['Time of Occurrence'] = pd.to_datetime(crimeDataset['Time of Occurrence'], format='%H:%M').dt.strftime('%I:%M %p')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sex_counts = crimeDataset['Victim Sex'].value_counts()
palette = sns.light_palette("red", len(sex_counts))

def autopct_func(pct):
    return '{:.1f}%\n({:.0f})'.format(pct, pct/100*sum(sex_counts))

sex_counts.plot.pie(autopct=autopct_func, startangle=90, colors = palette)
plt.axis('equal')
plt.legend(labels=['Male', 'Female', 'Unknown'])
plt.show()
```

```python
corr = crimeDataset.corr()
plt.figure(figsize=(9, 9))
corrMat = plt.matshow(corr, fignum = 1)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.gca().xaxis.tick_bottom()
plt.colorbar(corrMat)
plt.title(f'Correlation Matrix ', fontsize=15)
plt.show()
```

```python
corr_matrix = crimeDataset.corr()

# Set the figure size
fig, ax = plt.subplots(figsize=(8, 5))

colors = ["#017f8d", "#038c95", "#05a1a9", "#09b6bd", "#39cfd5", "#8ed8e8", "#c3e3ea", "#e9f8fb"]


# Create a correlation matrix plot
sns.heatmap(corr_matrix, annot=True, cmap = colors, ax=ax)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Group the data by date of occurrence and count the number of occurrences
crime_counts = crimeDataset.groupby('Date of Occurrence')['Crime Code'].count()

# Create a line chart
fig, ax = plt.subplots()
crime_counts.plot(kind='line', color='teal', figsize = (18,8))


# Customize the plot
ax.set_title('Trend of Crime Over Time')
ax.set_xlabel('Date of Occurrence')
ax.set_ylabel('Number of Occurrences')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

crime_counts = pd.pivot_table(crimeDataset, values='Crime Code', index='Area Name', columns='Crime Type- Part 1/2', aggfunc='count')

# Create a bar chart
fig, ax = plt.subplots()
colors = ["#017f8d","#39cfd5"]
crime_counts.plot(kind='bar', ax=ax, color = colors)

# Customize the plot
ax.set_title('Frequency of Different Types of Crimes by Area')
ax.set_xlabel('Area')
ax.set_ylabel('Number of Occurrences')

# Display the plot
plt.show()
```

```python
import plotly.express as px

# Assuming you have already defined and processed your data
crime_counts = crimeDataset.groupby(['Latitude', 'Longitude'])['Crime Code'].count().reset_index()

# Create a world heat map
fig = px.density_mapbox(crime_counts, lat='Latitude', lon='Longitude', z='Crime Code',
                        radius=10, center=dict(lat=0, lon=180),
                        zoom=0, mapbox_style='carto-positron',
                        hover_name='Crime Code',
                        hover_data={'Latitude': False, 'Longitude': False, 'Crime Code': True},
                        color_continuous_scale='Teal') # Set the color scale to 'Teal'

# Customize the plot
fig.update_layout(title='World Heat Map of Crime Occurrences')

# Display the plot
fig.show()
```
